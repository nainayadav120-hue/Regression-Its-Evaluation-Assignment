```python
#1. What is Simple Linear Regression?
""" Simple Linear Regression (SLR) is a statistical method used to model the relationship between a single independent variable (feature) and a continuous dependent variable (target).
It aims to find the best-fitting linear line that predicts the value of the target variable based on the value of the feature.

The equation for SLR is: y = β0 + β1x + ε
where:
- y is the dependent variable (target)
- x is the independent variable (feature)
- β0 is the intercept
- β1 is the slope coefficient
- ε is the error term
The goal is to estimate the coefficients (β0 and β1) that minimize the difference between predicted and actual values."""
```

```python
#2. What are the key assumptions of Simple Linear Regression?
""" The key assumptions of Simple Linear Regression (SLR) are:
1. Linearity: The relationship between the independent variable and the dependent variable is linear.
2. Independence: Observations are independent of each other.
3. Homoscedasticity: The variance of the residuals is constant across all levels of the independent variable.
4. Normality: The residuals are normally distributed.
5. No multicollinearity: Not applicable in SLR (since there's only one predictor), but in general, it means predictors shouldn't be highly correlated with each other.
These assumptions need to be met for the model to provide reliable and accurate predictions.
Violations of these assumptions can lead to biased or incorrect conclusions."""
```

```python
#3. What is heteroscedasticity, and why is it important to address in regression models?
""" Heteroscedasticity refers to the condition in which the variance of the residuals or error terms in a regression model is not constant across all levels of the independent variable(s).
In other words, the spread of the residuals changes as the values of the independent variable(s) change.
Heteroscedasticity is important to address in regression models because it can lead to:
1. Inefficient estimates: Heteroscedasticity can result in biased standard errors, which can affect the accuracy of hypothesis tests and confidence intervals.
2. Incorrect conclusions: Ignoring heteroscedasticity can lead to incorrect conclusions about the relationships between variables, potentially resulting in type I or type II errors.
3. Model misspecification: Heteroscedasticity can indicate model misspecification, such as omitted variables or incorrect functional form.
Addressing heteroscedasticity can involve:
- Data transformation (e.g., log transformation)
- Using robust standard errors
- Weighted least squares regression
- Modeling the variance structure (e.g., generalized estimating equations)
By addressing heteroscedasticity, you can improve the reliability and accuracy of your regression model."""
```

```python
#4. What is Multiple Linear Regression?
""" Multiple Linear Regression (MLR) is a statistical method that models the relationship between a continuous dependent variable (target) and
multiple independent variables (features).
It extends Simple Linear Regression by incorporating multiple predictors to predict the outcome variable.
The equation for MLR is: y = β0 + β1x1 + β2x2 + … + βnxn + ε
where:
- y is the dependent variable (target)
- x1, x2, …, xn are the independent variables (features)
- β0 is the intercept
- β1, β2, …, βn are the coefficients for each feature
- ε is the error term
MLR aims to estimate the coefficients (β) that minimize the difference between predicted and actual values,
while controlling for the effects of all other predictors in the model."""
```

```python
#5. What is polynomial regression, and how does it differ from linear regression?
""" Polynomial Regression is a type of regression analysis where the relationship between the independent variable \( x \) and the dependent variable \( y \) is modeled as a polynomial of degree \( n \). Unlike linear regression, which fits a straight line to the data, polynomial regression fits a curve, allowing it to capture non-linear relationships.

Key Differences from Linear Regression:1. Linear vs. Non-linear:
    - Linear Regression: Assumes a straight-line relationship: \( y = \beta_0 + \beta_1 x + \epsilon \).
    - Polynomial Regression: Models the relationship as a polynomial: \( y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \dots + \beta_n x^n + \epsilon \).

2. Model Complexity:
    - Linear regression is simpler and works well for linear relationships.
    - Polynomial regression can fit more complex, curved relationships by adding higher-degree terms.

3. Overfitting:
    - Polynomial regression is more prone to overfitting, especially with high-degree polynomials, as it can fit the noise in the data.

4. Interpretability:
    - Linear regression coefficients are easier to interpret (a unit increase in \( x \) leads to a \(\beta\) unit increase in \(y\)).
    - Polynomial regression coefficients are harder to interpret directly due to the non-linear nature.

When to Use Polynomial Regression:- When the relationship between \( x \) and \( y \) is clearly non-linear.
- When you suspect that the effect of \( x \) on \( y \) changes direction (e.g., a U-shaped or inverted U-shaped relationship).
Example:A quadratic regression (a type of polynomial regression) might look like:
\[ y = \beta_0 + \beta_1 x + \beta_2 x^2 + \epsilon \]
In summary, polynomial regression is a flexible extension of linear regression that can model non-linear relationships by adding polynomial terms.
However, care must be taken to avoid overfitting and ensure the model generalizes well to new data."""
```

```python
#6. Implement a Python program to fit a Simple Linear Regression model to the following sample data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
#Plot the regression line over the data points.
#(Include your Python code and output in the code box below.)
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Get the coefficients
intercept = model.intercept_
slope = model.coef_[0]

print(f"Linear Regression Equation: Y = {slope:.2f}X + {intercept:.2f}")

# Predict Y values
Y_pred = model.predict(X)

# Plot the data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
#7. Fit a Multiple Linear Regression model on this sample data:
#● Area = [1200, 1500, 1800, 2000]
#● Rooms = [2, 3, 3, 4]
#● Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.
#(Include your Python code and output in the code box below.)
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define the features (X) and target (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients
intercept = model.intercept_
coefficients = model.coef_

print("Multiple Linear Regression Equation: Price = {:.2f} + {:.2f}*Area + {:.2f}*Rooms".format(intercept, coefficients[0], coefficients[1]))

# Calculate VIF for each feature
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns

print("\nVIF Results:")
print(vif)

# Interpretation:
# A VIF value of 1 indicates no multicollinearity,
# VIF between 1 and 5 indicates moderate multicollinearity,
# and VIF above 5 or 10 indicates high multicollinearity.
```

```python
#8. Implement polynomial regression on the following data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve.
#(Include your Python code and output in the code box below.)
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create a Polynomial Features object of degree 2
poly_features = PolynomialFeatures(degree=2)

# Transform the data into polynomial features
X_poly = poly_features.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Get the coefficients
coefficients = model.coef_
intercept = model.intercept_

print("Polynomial Regression Equation: Y = {:.2f} + {:.2f}*X + {:.2f}*X^2".format(intercept, coefficients[1], coefficients[2]))

# Generate X values for plotting the curve
X_curve = np.linspace(0, 6, 100).reshape(-1, 1)
X_curve_poly = poly_features.transform(X_curve)

# Predict Y values for the curve
Y_curve = model.predict(X_curve_poly)

# Plot the data points and the polynomial curve
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_curve, Y_curve, color='red', label='Polynomial Curve (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()
```

```python
#9. Create a residuals plot for a regression model trained on this data:
#● X = [10, 20, 30, 40, 50]
#● Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.
#(Include your Python code and output in the code box below.)
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals vs fitted values
plt.scatter(Y_pred, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

# Assess heteroscedasticity:
# If the residuals are randomly scattered around the horizontal axis with no pattern,
# it suggests homoscedasticity (constant variance).
# If the residuals show a pattern (e.g., cone-shaped or fan-shaped),
# it may indicate heteroscedasticity (non-constant variance).
```

```python
#10.  Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location.
#However, you detect heteroscedasticity and multicollinearity in your regressionnmodel.
#Explain the steps you would take to address these issues and ensure a robust model.
""" Addressing Heteroscedasticity and Multicollinearity in Regression Models

When building a regression model for predicting house prices, heteroscedasticity and multicollinearity can lead to inaccurate predictions and unreliable statistical inferences. Here's a step-by-step approach to addressing these issues:

Heteroscedasticity1. Detect Heteroscedasticity:
    - Use visual methods (residual plots, scatter plots) to check for patterns in residuals.
    - Perform statistical tests like the Breusch-Pagan test or White's test.

2. Address Heteroscedasticity:
    - Data Transformation: Apply transformations like log or square root to the dependent variable (house price) to stabilize variance.
    - Weighted Least Squares (WLS): Assign weights to observations based on the variance of residuals.
    - Robust Standard Errors: Use robust standard errors (e.g., Huber-White standard errors) to account for heteroscedasticity.

Multicollinearity1. Detect Multicollinearity:
    - Calculate Variance Inflation Factor (VIF) for each feature.
    - Check for high correlation coefficients between features.

2. Address Multicollinearity:
    - Feature Selection: Remove highly correlated features or select the most relevant ones based on domain knowledge.
    - Regularization Techniques: Use L1 (Lasso) or L2 (Ridge) regularization to reduce the impact of multicollinearity.
    - Principal Component Regression (PCR): Use principal components as features to reduce dimensionality and multicollinearity.
    - Partial Least Squares (PLS) Regression: Use PLS regression to reduce dimensionality and multicollinearity.

Model Evaluation and Validation1. Cross-Validation: Use techniques like k-fold cross-validation to evaluate the model's performance on unseen data.
2. Metrics: Monitor metrics like Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared to assess the model's performance.
3. Residual Analysis: Continuously monitor residual plots to ensure that heteroscedasticity and multicollinearity are under control."""
#Implementation in Python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load data
data = pd.DataFrame({
    'area': [1200, 1500, 1800, 2000],
    'rooms': [2, 3, 3, 4],
    'price': [250000, 300000, 320000, 370000]
})

# Define features (X) and target (y)
X = data[['area', 'rooms']]
y = data['price']

# Calculate VIF
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns

# Address multicollinearity using Ridge regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X, y)

# Predict and evaluate
y_pred = ridge_model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Use robust standard errors to address heteroscedasticity
import statsmodels.api as sm

X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit(cov_type='HC1')
print(model_sm.summary())
```
